```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('exercise.csv', delimiter=';')
d = np.array(df)

print(df.to_markdown())
print(d.shape)
```

```python
x1 = d[:, 0] # forecast 1
x2 = d[:, 1] # forecast 2
y = d[:, 2]  # actual values

# RMSE for the two forecasts
SE_1 = (x1 - y) ** 2
RMSE_1 = np.sqrt(np.mean(SE_1))

SE_2 = (x2 - y) ** 2
RMSE_2 = np.sqrt(np.mean(SE_2))

# make table
table = pd.DataFrame({
    'Forecast 1': [RMSE_1],
    'Forecast 2': [RMSE_2]
})
print(table.to_markdown())
```

```python
# MAE for the two forecasts
AE_1 = np.abs(x1 - y)
MAE_1 = np.mean(AE_1)

AE_2 = np.abs(x2 - y)
MAE_2 = np.mean(AE_2)

# make table
table = pd.DataFrame({
    'Forecast 1': [MAE_1],
    'Forecast 2': [MAE_2]
})
print(table.to_markdown())
```

```python
# log score for the two forecasts
LogS = lambda f, y: -1 * (y * np.log(f) + (1 - y) * np.log(1 - f))

logS_1 = LogS(x1, y)
logS_2 = LogS(x2, y)

# make table
table = pd.DataFrame({
    'Forecast 1': [np.mean(logS_1)],
    'Forecast 2': [np.mean(logS_2)]
})
print(table.to_markdown())
```

```python
plt.plot(x1, label='Forecast 1')
plt.plot(x2, label='Forecast 2')
plt.plot(y, label='Actual')
plt.legend()
plt.show()
```
